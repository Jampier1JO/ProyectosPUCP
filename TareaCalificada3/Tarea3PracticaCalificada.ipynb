```python
# Direccion que apunta directamente al archivo countryCODES.xlsx
fallecidos_path = append_data_path + "/fallecidosCOVID.csv"
```

```python
#Leemos el archivo CSV en Pandas
fallecidos_df = pd.read_csv(fallecidos_path,sep=";")
```

```python
fallecidos_df
```

```python
# Convertir FECHA_FALLECIMIENTO a datetime
fallecidos_df['FECHA_FALLECIMIENTO'] = pd.to_datetime(fallecidos_df['FECHA_FALLECIMIENTO'], format='%Y%m%d')
```

```python
# Agrupamos por mes y aÃ±o, y contamos los fallecidos
fallecidos_df['AÃ±o-Mes'] = fallecidos_df['FECHA_FALLECIMIENTO'].dt.to_period('M')  # Crear una nueva columna para agrupar por mes y aÃ±o
reporte_mensual = fallecidos_df.groupby('AÃ±o-Mes').size().reset_index(name='Fallecidos')
reporte_mensual
```
